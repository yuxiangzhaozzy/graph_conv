```python
import requests
import json

def getjson(loc):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'} 
    pa = {
        'q': '冠捷大厦', 
        'region': loc,
        'scope': '2',
        'page_size': 20,
        'page_num': 0, 
        'output': 'json',
        'ak': '7hlbSyjePqmVEl1NIPujxYu4wjYhFGXr'
    }
    r = requests.get("http://api.map.baidu.com/place/v2/search", params=pa, headers= headers)
    decodejson = json.loads(r.text)
    return decodejson

getjson('北京')
```
